```python
# First import packages

import pandas as pd
import numpy as np
import pdb
import seaborn as sns
from matplotlib import pyplot as plt
import os
```

```python
print("Pandas version {}".format(pd.__version__))
print("Numpy version {}".format(np.__version__))
print("Seaborn version {}".format(sns.__version__))
```

```python
#Set the color for each projection population

colors = sns.color_palette('muted', 10)

PFC_color = "#FFDB6E"
NAc_color = "#BF43CC"
LH_color = "#87FFE1"
```

```python
#Change to directory containing .csv file
file_dir = '.'
os.chdir(file_dir)
```

```python
#Load the .csv file for coronal CTB data
df = pd.read_csv("coronal_CTBLabelled_dataset.csv")
df.head(5)
```

```python
#See the column names containing the types of data
df.columns
```

```python
#Filter dataframe by projection type
df_subset_PFC = df[df.Projection == 'PFC']
df_subset_NAc = df[df.Projection == 'NAc']
df_subset_LH = df[df.Projection == 'LH']
```

```python

#Plot the data in 3D

def white_bg_3d(ax):

    # make the grid lines transparent
    ax.xaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.yaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    ax.zaxis._axinfo["grid"]['color'] =  (1,1,1,0)
    
fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(projection='3d')

white_bg_3d(ax)

ax.scatter(df_subset_PFC['AP'], 
           df_subset_PFC['ML'] ,
           df_subset_PFC['DV'], alpha = 0.1, label = 'PFC', s = 8, color = PFC_color)

ax.scatter(df_subset_NAc['AP'], 
           df_subset_NAc['ML'] ,
           df_subset_NAc['DV'], alpha = 0.1, label = 'NAc', s = 8, color = NAc_color)

ax.scatter(df_subset_LH['AP'], 
           df_subset_LH['ML'] ,
           df_subset_LH['DV'], alpha = 0.1, label = 'LH', s = 8, color = LH_color)

ax.set_xlabel("AP")
ax.set_ylabel("ML")
ax.set_zlabel("DV")

plt.legend(markerscale = 3)

plt.show()
```

```python
#Load the .csv file for TRIO
df_trio = pd.read_csv("PFC_TRIO_cellcounts.csv")
df_trio.head(5)
```

```python
#Plot the data for PFC TRIO
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(projection='3d')

white_bg_3d(ax)

ax.scatter(df_trio['AP'], 
           df_trio['ML'] ,
           df_trio['DV'], alpha = 0.1, s = 8, color = df_trio['color'], label="TRIO")

ax.set_xlabel("AP")
ax.set_ylabel("ML")
ax.set_zlabel("DV")

ax.set_ylim(4,-4)
ax.set_xlim(-7,1)
ax.set_zlim(-8,-2)

plt.legend(markerscale = 3)
ax.set_aspect('equal')

plt.show()
```
